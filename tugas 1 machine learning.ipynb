```python
import pandas as pd
data = pd.read_csv('/content/dataset penyakit kanker payudara.csv')

data.info()
```

```python
data.head(50)
```

```python
data.drop('jenis_kelamin',axis=1, inplace=True)
data.drop('id',axis=1, inplace=True)
```

```python
data.head()
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()

data['pembengkakan_kelenjar'] = enc.fit_transform(data['pembengkakan_kelenjar'].values)
data['nyeri'] = enc.fit_transform(data['nyeri'].values)
data['perubahan_kulit'] = enc.fit_transform(data['perubahan_kulit'].values)
data['perubahan_bentuk'] = enc.fit_transform(data['perubahan_bentuk'].values)
data['terkena_kanker'] = enc.fit_transform(data['terkena_kanker'].values)
```

```python
#memisakhan atribut dengan label dimana x itu atribut dan y itu label
X = data[['perubahan_kulit','pembengkakan_kelenjar','nyeri']]
y = data['terkena_kanker']


#membagi dataset menjadi data latih dan data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.1, random_state=50)

from sklearn.tree import DecisionTreeClassifier

#membuat model decision tree
tree_model= DecisionTreeClassifier()

#melatih model menggunakan data latih dengan fungsi fit
tree_model= tree_model.fit(X_train,y_train)

#evaluasi model untuk meguji seberapa baik model yang digunakan dan akurasi

from sklearn.metrics import accuracy_score

y_pred =  tree_model.predict(X_test)

accuracy_score =round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ',accuracy_score)
```

```python
#prediksi model dengna tree model yang sdh dibuat dengan varibel ([''perubahan_kulit','pembengkakan_kelenjar','nyeri''])

print(tree_model.predict([[6.2, 3.3, 2.0]])[0])

print(tree_model.predict([[3.2, 3.3, 2.0]])[0])

print(tree_model.predict([[3.2, 3.3, 2.0]])[0])
```
